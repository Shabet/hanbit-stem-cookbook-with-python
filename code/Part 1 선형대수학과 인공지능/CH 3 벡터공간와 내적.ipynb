```python
import numpy as np

# ì½”ì‹œ-ìŠˆë°”ë¥´ì¸  ë¶€ë“±ì‹ì„ í™•ì¸í•˜ëŠ” í•¨ìˆ˜
# np.abs() : ()ì˜ ì ˆëŒ“ê°’
def check_Cauchy_Schwarz_inequality(vec1, vec2):
    innerprod_vecs = np.sum(vec1*vec2)
    norm_vec1 = np.sqrt(np.sum(vec1*vec1))
    norm_vec2 = np.sqrt(np.sum(vec2*vec2))
    check_CS_inequal = np.abs(innerprod_vecs) <= norm_vec1*norm_vec2
    return check_CS_inequal

# xì™€ yë¡œ í™•ì¸
x = np.array([1,-1,3])
y = np.array([1,1,3])
print("Cauchy-Schwarz inequality is", check_Cauchy_Schwarz_inequality(x,y))
```

```python
from sympy import *
import numpy as np

# ë¬¸ì œ (a)
# symbols('x') : xë¥¼ ë¯¸ì§€ìˆ˜ë¡œ ì„ ì–¸
# diff(f, x) : í•¨ìˆ˜ fë¥¼ xë¡œ ë¯¸ë¶„
x, y, z = symbols('x y z')
f = x*y*z

df_dx = diff(f, x)
df_dy = diff(f, y)
df_dz = diff(f, z)

gradient_f_1 = np.array([[df_dx], [df_dy], [df_dz]])
print("(a) gradient of f = \n", gradient_f_1)
print("\n")

# ë¬¸ì œ (b)
x, y, z = symbols('x y z')
f = x**2*y**3+z**2+exp(x*y)

df_dx = diff(f, x)
df_dy = diff(f, y)
df_dz = diff(f, z)

gradient_f_1 = np.array([[df_dx], [df_dy], [df_dz]])
print("(b) gradient of f = \n", gradient_f_1)
```

```python
from sympy import *
import numpy as np

x, y, z = symbols('x y z')
F_1 = x-y+z**2
F_2 = x*cos(y)
F_3 = y**2*exp(3*z)

# F1, F2, F3ë¥¼ xë¡œ ë¯¸ë¶„
df1_dx = diff(F_1, x)
df2_dx = diff(F_2, x)
df3_dx = diff(F_3, x)

# F1, F2, F3ë¥¼ yë¡œ ë¯¸ë¶„
df1_dy = diff(F_1, y)
df2_dy = diff(F_2, y)
df3_dy = diff(F_3, y)

# F1, F2, F3ë¥¼ zë¡œ ë¯¸ë¶„
df1_dz = diff(F_1, z)
df2_dz = diff(F_2, z)
df3_dz = diff(F_3, z)

Jacobian_F = np.array([[df1_dx, df2_dx, df3_dx], [df1_dy, df2_dy, df3_dy], [df1_dz, df2_dz, df3_dz]])
print("Jacobian of F = \n", Jacobian_F)
```

3.1 ì£¼ì–´ì§„ ë²¡í„° x, y, z,v, wì— ëŒ€í•˜ì—¬ ë‹¤ìŒì„ êµ¬í•˜ë¼.

```python
import numpy as np

# ë…¸ë¦„ì„ êµ¬í•˜ëŠ” í•¨ìˆ˜
def cal_norm(vec1):
    norm_vec = np.sqrt(np.sum(vec1*vec1))
    return norm_vec

# ë‚´ì ì„ êµ¬í•˜ëŠ” í•¨ìˆ˜
def cal_innerproduct(vec1, vec2):
    innerprod_vecs = np.sum(vec1*vec2)
    return innerprod_vecs

# í•´ë° ê±°ë¦¬ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜
def cal_hamming_dist(vec1, vec2):
    Num_vec = len(vec1)
    hamming_dist = 0
    for i in range(Num_vec):
        if vec1[i] != vec2[i]:
            hamming_dist = hamming_dist+1
    return hamming_dist

# ë§¨í•˜íŠ¼ ê±°ë¦¬ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜
def cal_Manhattan_dist(vec1, vec2):
    Manhattan_dist = np.sum(np.abs(vec1-vec2))
    return Manhattan_dist

# ë²¡í„° x, y, z, v, w
x = np.array([1,0,1,0])
y = np.array([0,2,2,0])
z = np.array([1,0,1,2])
v = np.array([1,1,1,1])
w = np.array([2,2,0,1])

# ë¬¸ì œ (a)
print("ë¬¸ì œ (a)")
print("Norm of x = ", cal_norm(x))
print("Norm of y = ", cal_norm(y))
print("Norm of z = ", cal_norm(z))
print("\n")

# ë¬¸ì œ (b)
print("ë¬¸ì œ (b)")
print("Inner product of x and y = ", cal_innerproduct(x,y))
print("\n")

# ë¬¸ì œ (c)
print("ë¬¸ì œ (c)")
print("Inner product of z and x+y = ", cal_innerproduct(z, x+y))
print("\n")

# ë¬¸ì œ (d)
print("ë¬¸ì œ (d)")
print("Hamming distance of x and v = ", cal_hamming_dist(x,v))
print("\n")

# ë¬¸ì œ (e)
print("ë¬¸ì œ (e)")
print("Manhattan distance of v and w = ", cal_Manhattan_dist(v,w))
```

3.2 ë‹¤ìŒê³¼ ê°™ì€ ë‘ ë²¡í„° x, yê°€ ì´ë£¨ëŠ” ê°ì„ êµ¬í•˜ê³ , xì˜ y ìœ„ë¡œì˜ ì •ì‚¬ì˜ proj_y xë¥¼ êµ¬í•˜ë¼.

```python
import numpy as np

# ë‘ ë²¡í„°ê°€ ì´ë£¨ëŠ” ê°ì„ êµ¬í•˜ëŠ” í•¨ìˆ˜
# np.linalg.norm(vec) : vecì˜ ë…¸ë¦„
# np.arctan() : tan()ì˜ ì—­í•¨ìˆ˜ ê°’

def angle_to_vectors(vec_1, vec_2):
    vec_1_norm = np.linalg.norm(vec_1)
    vec_2_norm = np.linalg.norm(vec_2)
    vec_1_2_dot = np.dot(vec_1.T, vec_2)
    angle = np.arctan(vec_1_2_dot/(vec_1_norm*vec_2_norm))*360/np.pi
    return angle

# ì •ì‚¬ì˜ì„ êµ¬í•˜ëŠ” í•¨ìˆ˜
def Proj_onto(vec_1, vec_2):
    vec_2_1_dot = np.dot(vec_2.T, vec_1)
    vec_2_dot = np.dot(vec_2.T, vec_2)
    proj_vec = (vec_2_1_dot/vec_2_dot)*vec_2
    return proj_vec

# ë¬¸ì œ (a)
x = np.array([[1], [0]])
y = np.array([[4], [4]])
angle = angle_to_vectors(x, y)
projxy = Proj_onto(y, x)
print("ë¬¸ì œ (a)")
print("xì™€ yê°€ ì´ë£¨ëŠ” ê° = ", angle)
print("xì˜ y ìœ„ë¡œì˜ ì •ì‚¬ì˜ = \n", projxy)
print("\n")

# ë¬¸ì œ (b)
x = np.array([[1], [2], [5]])
y = np.array([[1], [-1], [2]])
angle = angle_to_vectors(x, y)
projxy = Proj_onto(y, x)
print("ë¬¸ì œ (b)")
print("xì™€ yê°€ ì´ë£¨ëŠ” ê° = ", angle)
print("xì˜ y ìœ„ë¡œì˜ ì •ì‚¬ì˜ = \n", projxy)
print("\n")

# ë¬¸ì œ (c)
x = np.array([[0], [-1], [2], [2]])
y = np.array([[1], [1], [3], [2]])
angle = angle_to_vectors(x, y)
projxy = Proj_onto(y, x)
print("ë¬¸ì œ (c)")
print("xì™€ yê°€ ì´ë£¨ëŠ” ê° = ", angle)
print("xì˜ y ìœ„ë¡œì˜ ì •ì‚¬ì˜ = \n", projxy)
```

3.3 ë‹¤ìŒê³¼ ê°™ì´ ì£¼ì–´ì§„ ì  Pë¥¼ í¬í•¨í•˜ë©° ë²•ì„ ë²¡í„°ê°€ nì¸ í‰ë©´ì´ ìˆë‹¤ê³  í•˜ì. ë‚´ì ì„ ì´ìš©í•˜ì—¬ ì´ í‰ë©´ê³¼ ì  P'ì‚¬ì´ì˜ ê±°ë¦¬ë¥¼ êµ¬í•˜ë¼.

```python
import numpy as np

# í‰ë©´ê³¼ ì  ì‚¬ì´ì˜ ê±°ë¦¬ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜
# np.absolute() : np.abs()ì™€ ê°™ì€ í•¨ìˆ˜
def dist_Pt_to_Pl(P, N_vec, PP) :
    num = np.dot((PP-P).T, N_vec)
    N_norm = np.linalg.norm(N_vec)
    dist = np.absolute(num)/N_norm
    return dist

# ë¬¸ì œ (a)
print("ë¬¸ì œ (a)")
P = np.array([1, 9, 8])
N = np.array([7, 5, 2])
PP = np.array([5, 1, 9])
print("í‰ë©´ê³¼ ì  ì‚¬ì´ì˜ ê±°ë¦¬ = ", dist_Pt_to_Pl(P, N, PP))
print("\n")

# ë¬¸ì œ (b)
print("ë¬¸ì œ (b)")
P = np.array([1, 4, 5, 0])
N = np.array([4, 1, 2, 0])
PP = np.array([17, 9, 4, 11])
print("í‰ë©´ê³¼ ì  ì‚¬ì´ì˜ ê±°ë¦¬ = ", dist_Pt_to_Pl(P, N, PP))
```

```python
import sympy as sym

# 함수 정의
x, y = sym.symbols('x y') # x, y를 미지수로 선언
f = (2*x*(y-1)**2)/(x**2+(y-1)**2) # 함수 f
# x, y에 들어갈 입력값
# x는 0에, y는 1에 가깝게
x_val = [0.1, 0.01, 0.001]
y_val = [0.9, 0.99, 0.999]

# epsilon 설정
epsilon = 0.01

# 함수.subs() : 함수에 값을 대입하여 값을 반환하는 함수
print("x = 0.1, y= 0.9일 때")
f_1 = f.subs([(x, x_val[0]), (y, y_val[0])])
print("f(",x_val[0],",",y_val[0],")=", f_1)
print("f(",x_val[0],",",y_val[0],")이 epsilon보다 작은가?", abs(f_1 - 0) < epsilon)
print("\n")

print("x = 0.01, y= 0.09일 때")
f_2 = f.subs([(x, x_val[1]), (y, y_val[1])])
print("f(",x_val[1],",",y_val[1],")=", f_2)
print("f(",x_val[1],",",y_val[1],")이 epsilon보다 작은가?", abs(f_2 - 0) < epsilon)
print("\n")

print("x = 0.001, y= 0.009일 때")
f_3 = f.subs([(x, x_val[2]), (y, y_val[2])])
print("f(",x_val[2],",",y_val[2],")=", f_3)
print("f(",x_val[2],",",y_val[2],")이 epsilon보다 작은가?", abs(f_3 - 0) < epsilon)
```

```python
import sympy as sym
import numpy as np

x, y, L= sym.symbols('x y L') # x, y, L을 미지수로 선언
f = x**2+ y**2
g = 3*x**2 + 4*y**2 - 5

df_dx = sym.diff(f, x) # x로 미분
df_dy = sym.diff(f, y) # y로 미분

dg_dx = sym.diff(g, x) # x로 미분
dg_dy = sym.diff(g, y) # y로 미분

# 수식.coeff() : 수식의 특정항의 계수를 반환하는 함수
print("f(x,y) = ", f)
print("g(x,y) = ", g)
print("x = ", dg_dx*L / df_dx.coeff(x,1))
print("y = ", dg_dy*L / df_dy.coeff(y,1))
print("\n")

print("경우 1) x = 0")
g_y = g.subs([(x, 0)])
print("g(0, y) = ", g_y)
g_y_sol = sym.solve(sym.Eq(g_y, 0), y)
print("y = ", g_y_sol)
print("f(x, ",g_y_sol[0], ") = ", f.subs([(x, 0), (y, g_y_sol[0])]))
print("f(x, ",g_y_sol[1], ") = ", f.subs([(x, 0), (y, g_y_sol[1])]))
print("\n")

print("경우 2) x != 0")
g_x = g.subs([(y, 0)])
print("g(x, 0) = ", g_x)
g_x_sol = sym.solve(sym.Eq(g_x, 0), x)
print("x = ", g_x_sol)
print("f(" ,g_x_sol[0], ", 0 ) = ", f.subs([(x, g_x_sol[0]), (y, 0)]))
print("f(" ,g_x_sol[1], ", 0 ) = ", f.subs([(x, g_x_sol[1]), (y, 0)]))
print("\n")

print("f의 최댓값 = ", f.subs([(x, g_x_sol[0]), (y, 0)]), "이고, ")
print("이때 (x, y) = ",(g_x_sol[0], 0), ",", (g_x_sol[1], 0), "이다." )
print("f의 최솟값 = ", f.subs([(x, 0), (y, g_y_sol[0])]), "이고, ")
print("이때 (x, y) = ",(0, g_y_sol[0]), ",", (0, g_y_sol[1]), "이다." )
```

```python
import numpy as np
import sympy as sym
# 어떤 정부호 행렬인지 확인하는 함수
def check_definite_matrix(func_f):
    import sympy as sym
    x, y = sym.symbols('x y')
    first_eval = func_f.subs([(x, 1), (y, 0)])
    second_eval = func_f.subs([(x, 0), (y, 1)])
    third_eval = func_f.subs([(x, -1), (y, 0)])
    fourth_eval = func_f.subs([(x, 0), (y, -1)])
    if (first_eval > 0) & (second_eval > 0) & (third_eval > 0) & (fourth_eval > 0) :
        Mat_name = 'positive definite matrix'
        return Mat_name
    elif (first_eval < 0) & (second_eval < 0) & (third_eval < 0) & (fourth_eval < 0) :
        Mat_name = 'negative definite matrix'
        return Mat_name

# 문제 (a)
A = np.array([[1,0],[0,1]])
x, y = sym.symbols('x y')
Poly_Mat_A = (sym.Matrix([x, y]).T * sym.Matrix(A)) * sym.Matrix([x, y])
Poly_A = np.array(Poly_Mat_A)
print("문제 (a)")
print('x^T*A*x = ', Poly_A[0][0])
print('행렬 A는', check_definite_matrix(Poly_A[0][0]), '이다.')
print('\n')

# 문제 (b)
B = np.array([[-1,1],[1,-2]])
x, y = sym.symbols('x y')
Poly_Mat_B = (sym.Matrix([x, y]).T * sym.Matrix(B)) * sym.Matrix([x, y])
Poly_B = np.array(Poly_Mat_B)
print("문제 (b)")
print('x^T*B*x = ', Poly_B[0][0])
print('행렬 B는', check_definite_matrix(Poly_B[0][0]), '이다.')
```

9.1 [정의 9-10]을 프로그래밍하여 다음 행렬이 어떤 정부호 행렬인지 판별하라.

```python
import sympy as sym
import numpy as np

# 어떤 정부호 행렬인지 확인하는 함수
def check_definite_matrix(func_f):
    import sympy as sym
    x, y = sym.symbols('x y' )
    first_eval = func_f.subs([(x, 1), (y, 0)])
    second_eval = func_f.subs([(x, 0), (y, 1)])
    third_eval = func_f.subs([(x, -1), (y, 0)])
    fourth_eval = func_f.subs([(x, 0), (y, -1)])
    if (first_eval > 0) & (second_eval > 0) & (third_eval > 0) & (fourth_eval > 0) :
        Mat_name = 'positive definite matrix'
        return Mat_name
    elif (first_eval >= 0) & (second_eval >= 0) & (third_eval >= 0) & (fourth_eval >= 0) :
        Mat_name = 'positive semi-definite matrix'
        return Mat_name
    elif (first_eval < 0) & (second_eval < 0) & (third_eval < 0) & (fourth_eval < 0) :
        Mat_name = 'negative definite matrix'
        return Mat_name
    elif (first_eval <= 0) & (second_eval <= 0) & (third_eval >= 0) & (fourth_eval >= 0) :
        Mat_name = 'negative semi-definite matrix'
        return Mat_name
    else:
        Mat_name = 'indefinite matrix'
        return Mat_name

# 문제 (a)
A = np.array([[2,2],[3,1]])
x, y = sym.symbols('x y' )
Poly_Mat_A = (sym.Matrix([x, y]).T * sym.Matrix(A)) * sym.Matrix([x, y])
Poly_A = np.array(Poly_Mat_A)
print("문제 (a)")
print('x^T*A*x = ', Poly_A[0][0])
print('행렬 A는', check_definite_matrix(Poly_A[0][0]), '이다.')
print('\n')

# 문제 (b)
B = np.array([[-2,5],[7,-16]])
x, y = sym.symbols('x y' )
Poly_Mat_B = (sym.Matrix([x, y]).T * sym.Matrix(B)) * sym.Matrix([x, y])
Poly_B = np.array(Poly_Mat_B)
print("문제 (b)")
print('x^T*B*x = ', Poly_B[0][0])
print('행렬 B는', check_definite_matrix(Poly_B[0][0]), '이다.')
print('\n')

# 문제 (c)
C = np.array([[-2,1],[-2,1]])
x, y = sym.symbols('x y' )
Poly_Mat_C = (sym.Matrix([x, y]).T * sym.Matrix(C)) * sym.Matrix([x, y])
Poly_C = np.array(Poly_Mat_C)
print("문제 (c)")
print('x^T*C*x = ', Poly_C[0][0])
print('행렬 C는', check_definite_matrix(Poly_C[0][0]), '이다.')
print('\n')

# 문제 (d)
D = np.array([[1,5],[5,8]])
x, y = sym.symbols('x y' )
Poly_Mat_D = (sym.Matrix([x, y]).T * sym.Matrix(D)) * sym.Matrix([x, y])
Poly_D = np.array(Poly_Mat_D)
print("문제 (d)")
print('x^T*D*x = ', Poly_D[0][0])
print('행렬 D는', check_definite_matrix(Poly_D[0][0]), '이다.')
```

9.2 [정리 9-10]을 프로그래밍하여 다음 함수의 극대점, 극소점 또는 안장점을 구하라.

```python
import sympy as sym
import numpy as np

# 이변수함수의 헤시안 행렬을 계산하는 함수
def cal_Hessian(func_f):
    import sympy as sym
    import numpy as np
    
    x, y = sym.symbols('x y' )
    d2func_f_dx2 = sym.diff(func_f, x, 2) # x로 두 번 미분
    d2func_f_dyx = sym.diff(func_f, y, x) # y, x 순서로 미분
    d2func_f_dxy = sym.diff(func_f, x, y) # x, y 순서로 미분
    d2func_f_dy2 = sym.diff(func_f, y, 2) # y로 두 번 미분
    
    Hessian_func_f = np.array([[d2func_f_dx2, d2func_f_dyx], [d2func_f_dxy, d2func_f_dy2]])
    return Hessian_func_f

# 이변수함수의 그래디언트를 계산하는 함수
def cal_gradient(func_f):
    import sympy as sym
    import numpy as np
    
    x, y = sym.symbols('x y' )
    d2func_f_dx = sym.diff(func_f, x) # x로 두 번 미분
    d2func_f_dy = sym.diff(func_f, y) # y로 두 번 미분
    
    gradient_func_f = np.array([[d2func_f_dx], [d2func_f_dy]])
    return gradient_func_f

# 이변수함수의 그래디언트가 0인 점을 계산하는 함수
def cal_gradient_point(func_f) :
    import sympy as sym
    
    x, y = sym.symbols('x y' )
    grad_func_f = cal_gradient(func_f)
    gradient_Points = sym.solve([grad_func_f[0][0], grad_func_f[1][0]], x, y)
    return gradient_Points

# 이변수함수와 점을 이용하여 극점, 함숫값을 계산하는 함수
def cal_extreme_eval(func_f, eval_point):
    import sympy as sym
    import numpy as np
    
    x, y = sym.symbols('x y' )
    Hessian_func_f = cal_Hessian(func_f)
    Hessian_func_f_eval = np.zeros([2,2])
    Hessian_func_f_eval[0][0] = Hessian_func_f[0][0].subs([(x, eval_point[0]), 
                                                           (y, eval_point[1])])
    Hessian_func_f_eval[1][0] = Hessian_func_f[1][0].subs([(x, eval_point[0]),
                                                           (y, eval_point[1])])
    Hessian_func_f_eval[0][1] = Hessian_func_f[0][1].subs([(x, eval_point[0]),
                                                           (y, eval_point[1])])
    Hessian_func_f_eval[1][1] = Hessian_func_f[1][1].subs([(x, eval_point[0]),
                                                           (y, eval_point[1])])
    
    np_Hessian_func_f = np.array(Hessian_func_f_eval)
    Hessian_func_f_eig = np.linalg.eigvals(np_Hessian_func_f)
    det_Hessian_func_f_eig = np_Hessian_func_f[0][0]*np_Hessian_func_f[1][1]-np_Hessian_func_f[0][1]*np_Hessian_func_f[1][0]

    if np.all(Hessian_func_f_eig > 0) == True :
        eval_point_state = '극소점'
        func_f_point_eval = func_f.subs([(x, eval_point[0]), (y, eval_point[1])])
        return eval_point_state, func_f_point_eval
    elif np.all(Hessian_func_f_eig < 0) == True :
        eval_point_state = '극대점'
        func_f_point_eval = func_f.subs([(x, eval_point[0]), (y, eval_point[1])])
        return eval_point_state, func_f_point_eval
    elif (det_Hessian_func_f_eig < 0) == True :
        eval_point_state = '안장점'
        func_f_point_eval = func_f.subs([(x, eval_point[0]), (y, eval_point[1])])
        return eval_point_state, func_f_point_eval
    else:
        eval_point_state = '극소점, 극대점, 안장점이 모두 아닌 점'
        func_f_point_eval = func_f.subs([(x, eval_point[0]), (y, eval_point[1])])
        return eval_point_state, func_f_point_eval
    
# 문제 (a)
print("문제 (a)")
x, y = sym.symbols('x y') # x, y를 미지수로 선언
f = 9-2*x+4*y-x**2-4*y**2
Hessian_f = cal_Hessian(f)
print("Hessian of f = \n", Hessian_f)
print('\n')

gradient_f = cal_gradient(f)
print("gradient of f = \n", gradient_f)
print('\n')
test_Points = cal_gradient_point(f)
print("f의 그래디언트가 0인 점 = \n", test_Points)
print('\n')

# 첫 번째 점
first_point = (test_Points[x], test_Points[y])
print(first_point, '에서 구하기')
state, point_eval = cal_extreme_eval(f, first_point)
print(first_point, "은 ", state, "이다.")
print("f", first_point, "=", point_eval)
print('\n')
print('-----------------------------------------------------------------')

# 문제 (b)
print("문제 (b)")
x, y = sym.symbols('x y') # x, y를 미지수로 선언
f = x*y*(1-x-y)
Hessian_f = cal_Hessian(f)
print("Hessian of f = \n", Hessian_f)
print('\n')

gradient_f = cal_gradient(f)
print("gradient of f = \n", gradient_f)
print('\n')
test_Points = cal_gradient_point(f)
print("f의 그래디언트가 0인 점 = \n", test_Points)
print('\n')

# 첫 번째 점
first_point = test_Points[0]
print(first_point, '에서 구하기')
state, point_eval = cal_extreme_eval(f, first_point)
print(first_point, "은 ", state, "이다.")
print("f", first_point, "=", point_eval)
print('\n')

# 두 번째 점
second_point = test_Points[1]
print(second_point, '에서 구하기')
state, point_eval = cal_extreme_eval(f, second_point)
print(second_point, "은 ", state, "이다.")
print("f", second_point, "=", point_eval)
print('\n')

# 세 번째 점
third_point = test_Points[2]
print(third_point, '에서 구하기')
state, point_eval = cal_extreme_eval(f, third_point)
print(third_point, "은 ", state, "이다.")
print("f", third_point, "=", point_eval)
print('\n')

# 네 번째 점
fourth_point = test_Points[3]
print(fourth_point, '에서 구하기')
state, point_eval = cal_extreme_eval(f, fourth_point)
print(fourth_point, "은 ", state, "이다.")
print("f", fourth_point, "=", point_eval)
print('\n')
print('-----------------------------------------------------------------')

# 문제 (c)
print("문제 (c)")
x, y = sym.symbols('x y') # x, y를 미지수로 선언
f = y**2-2*y*sym.cos(x)
Hessian_f = cal_Hessian(f)
print("Hessian of f = \n", Hessian_f)
print('\n')

gradient_f = cal_gradient(f)
print("gradient of f = \n", gradient_f)
print('\n')
test_Points = cal_gradient_point(f)
print("f의 그래디언트가 0인 점 = \n", test_Points)
print('\n')

# 첫 번째 점
first_point = test_Points[0]
print(first_point, '에서 구하기')
state, point_eval = cal_extreme_eval(f, first_point)
print(first_point, "은 ", state, "이다.")
print("f", first_point, "=", point_eval)
print('\n')

# 두 번째 점
second_point = test_Points[1]
print(second_point, '에서 구하기')
state, point_eval = cal_extreme_eval(f, second_point)
print(second_point, "은 ", state, "이다.")
print("f", second_point, "=", point_eval)
print('\n')

# 세 번째 점
third_point = test_Points[2]
print(third_point, '에서 구하기')
state, point_eval = cal_extreme_eval(f, third_point)
print(third_point, "은 ", state, "이다.")
print("f", third_point, "=", point_eval)
print('\n')

# 네 번째 점
fourth_point = test_Points[3]
print(fourth_point, '에서 구하기')
state, point_eval = cal_extreme_eval(f, fourth_point)
print(fourth_point, "은 ", state, "이다.")
print("f", fourth_point, "=", point_eval)
```

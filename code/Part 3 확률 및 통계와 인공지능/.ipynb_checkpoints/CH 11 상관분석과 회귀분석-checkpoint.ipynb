# 11.1 상관분석

예제 11-1

```python
import numpy as np

# X, Y의 확률분포
Prop_X_Y = np.array([[1/8,1/4,0],[1/8,1/4,1/4]])

# 문제 (a)
print("문제 (a)")
X = np.array([1,2,3])
Prop_X = np.sum(Prop_X_Y, axis=0)
Exp_X = np.sum((X*Prop_X))
print("E(X) = ", Exp_X)
Exp_X_2 = np.sum((X**2)*(Prop_X))
Var_X = Exp_X_2 - (Exp_X**2)
print("V(X) = ", Var_X)
print("\n")

# 문제 (b)
print("문제 (b)")
Y = np.array([1,2])
Prop_Y = np.sum(Prop_X_Y, axis=1)
Exp_Y = np.sum((Y*Prop_Y))
print("E(Y) = ", Exp_Y)
Exp_Y_2 = np.sum((Y**2)*(Prop_Y))
Var_Y = Exp_Y_2 - (Exp_Y**2)
print("V(Y) = ", Var_Y)
print("\n")

# 문제 (c)
print("문제 (c)")
Exp_X_Y = 0
for i in range(len(X)):
    for j in range(len(Y)):
        cal_val = X[i]*Y[j]*Prop_X_Y[j][i]
        Exp_X_Y = Exp_X_Y + cal_val
print("E(XY) = ", Exp_X_Y)
print("\n")

# 문제 (d)
print("문제 (d)")
Cov_X_Y = Exp_X_Y - (Exp_X * Exp_Y)
print("Cov(X, Y) = ", Cov_X_Y)
print("\n")

# 문제 (e)
print("문제 (e)")
Var_X_Y = Var_X + Var_Y + (2*Cov_X_Y)
print("V(X+Y) = ", Var_X_Y)
```

예제 11-2 

```python
import numpy as np
import sympy as sym

# 확률변수 X, Y의 결합확률분포표 만들기
x, y = sym.symbols('x y')
Prop_P = 1/18*(x+y)
X = np.arange(3)
Y = np.arange(3)

Prop_X_Y = []
for i in range(len(X)):
    cal_X = []
    for j in range(len(Y)):
        cal_X_Y = Prop_P.subs([(x, X[i]), (y, Y[j])])
        cal_X.append(cal_X_Y)
    Prop_X_Y.append(cal_X)
Prop_X_Y = np.array(Prop_X_Y)

# 문제 (a)
print("문제 (a)")
print("결합확률분포표 = \n", Prop_X_Y)
print("\n")

# 문제 (b)
print("문제 (b)")
# 확률변수 X의 기댓값, 분산 구하기
Prop_X = np.sum(Prop_X_Y, axis=0)
Exp_X = np.sum((X*Prop_X))
Exp_X_2 = np.sum((X**2)*(Prop_X))
Var_X = Exp_X_2 - (Exp_X**2)

# 확률변수 Y의 기댓값, 분산 구하기
Prop_Y = np.sum(Prop_X_Y, axis=1)
Exp_Y = np.sum((Y*Prop_Y))
Exp_Y_2 = np.sum((Y**2)*(Prop_Y))
Var_Y = Exp_Y_2 - (Exp_Y**2)

# 확률변수 X, Y의 기댓값, 분산과 상관계수 구하기
Exp_X_Y = 0
for i in range(len(X)):
    for j in range(len(Y)):
        cal_val = X[i]*Y[j]*Prop_X_Y[j][i]
        Exp_X_Y = Exp_X_Y + cal_val

Cov_X_Y = Exp_X_Y - (Exp_X * Exp_Y)
Var_X_Y = Var_X + Var_Y + (2*Cov_X_Y)
Corr_X_Y = Cov_X_Y/((Var_X**(1/2))*(Var_Y**(1/2)))
print("Corr(X, Y) = ", Corr_X_Y)
```

# 11.2 회귀 분석

예제 11-3

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

# 행렬 A와 벡터 b 만들기
A = np.array([[1, 0],[1, 2],[1, 3],[1, 5]])
b = np.array([1,2,4,5])
AtA = np.dot(A.T,A)
AtA_inv = np.linalg.inv(AtA)

# 최적의 근사해 β 구하기
Beta = np.round(np.dot(np.dot(AtA_inv, A.T),b),2)

# 선형회귀식
x = sym.Symbol('x')
lin_reg = Beta[0]+Beta[1]*x
print("선형회귀식 = ", lin_reg)
print("\n")

# 데이터와 선형회귀식의 그래프 그리기
plot_x = np.arange(0, 5.1)
plot_y = Beta[0]+Beta[1]*plot_x
plt.figure(figsize=(12, 8))
plt.plot(plot_x, plot_y, 'k--', label="y=0.88+0.85x")
plt.scatter(A[:,1], b)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.grid()
plt.show()
```

예제 11-4 

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

# 행렬 A와 벡터 b 만들기
A = np.array([[1, 2, 2],[1, 3, 4],[1, 5, 5],[1, 7, 8],[1, 9, 9]])
b = np.array([7,9,10,20,15])

print("A = \n", A)
print("\n")
print("b = \n", b)
print("\n")

AtA = np.dot(A.T,A)
print("A^T * A = \n", AtA)
print("\n")

AtA_inv = np.linalg.inv(AtA)
print("(A^T * A)^(-1) = \n", AtA_inv)
print("\n")

# 최적의 근사해 β 구하기
Beta = np.round(np.dot(np.dot(AtA_inv, A.T),b),2)
print("beta = \n", Beta)
print("\n")

# 선형회귀식
x, y = sym.symbols('x y')
lin_reg = Beta[0]+Beta[1]*x + Beta[2]*y
print("선형회귀식 = ", lin_reg)
```

11.1 

```python
import numpy as np

# 확률변수와 확률을 이용해 기댓값을 계산하는 함수
def cal_Expected_val(rand_variable, probability_P):
    Expected_val = np.sum((rand_variable * probability_P))
    return Expected_val

# 확률변수와 확률을 이용해 분산을 계산하는 함수
def cal_Variance(rand_variable, probability_P):
    Expected_rand = cal_Expected_val(rand_variable, probability_P)
    Rand_2 = rand_variable**2
    Expected_rand_2 = cal_Expected_val(Rand_2, probability_P)
    Variance_val = Expected_rand_2 - (Expected_rand**2)
    return Variance_val

# 확률변수 X, Y와 확률분포표를 이용해 기댓값을 계산하는 함수
def cal_Expected_value_X_Y(rand_variable_X, rand_variable_Y, X_Y_probability_P):
    Expected_val_X_Y = 0
    for i in range(len(rand_variable_X)):
        for j in range(len(rand_variable_Y)):
            Joint_val = rand_variable_X[i]*rand_variable_Y[j]*X_Y_probability_P[j][i]
            Expected_val_X_Y = Expected_val_X_Y + Joint_val
    return Expected_val_X_Y

# 확률변수 X, Y와 확률분포표를 이용해 공분산을 계산하는 함수
def cal_Corvariance_X_Y(rand_variable_X, rand_variable_Y, X_Y_probability_P):
    Prop_rand_variable_X = np.sum(X_Y_probability_P, axis=0)
    Prop_rand_variable_Y = np.sum(X_Y_probability_P, axis=1)
    Expected_rand_X = cal_Expected_val(rand_variable_X, Prop_rand_variable_X)
    Expected_rand_Y = cal_Expected_val(rand_variable_Y, Prop_rand_variable_Y)
    Expected_rand_X_Y = cal_Expected_value_X_Y(rand_variable_X, rand_variable_Y, X_Y_probability_P)
    Cor_X_Y = Expected_rand_X_Y - (Expected_rand_X * Expected_rand_Y)
    return Cor_X_Y

# 확률변수 X, Y와 확률분포표를 이용해 V(X+Y)를 계산하는 함수
def cal_joint_variance_X_Y(rand_variable_X, rand_variable_Y, X_Y_probability_P):
    Prop_rand_variable_X = np.sum(X_Y_probability_P, axis=0)
    Prop_rand_variable_Y = np.sum(X_Y_probability_P, axis=1)
    Variance_rand_X = cal_Variance(rand_variable_X, Prop_rand_variable_X)
    Variance_rand_Y = cal_Variance(rand_variable_Y, Prop_rand_variable_Y)
    Corvariance_rand_X_Y = cal_Corvariance_X_Y(rand_variable_X, rand_variable_Y, X_Y_probability_P)
    Variance_rand_X_Y = Variance_rand_X + Variance_rand_Y + (2*Corvariance_rand_X_Y)
    return Variance_rand_X_Y

# 문제 (a)
print("문제 (a)")
X = np.array([0, 1])
Y = np.array([0, 1, 2])
Prop_X_Y = np.array([[1/6,1/3],[1/8,1/8],[1/24,5/24]])
Prop_X = np.sum(Prop_X_Y, axis=0)
Prop_Y = np.sum(Prop_X_Y, axis=1)

# 확률변수 X의 기댓값, 분산 구하기
Exp_X = cal_Expected_val(X, Prop_X)
print("E(X) = ", Exp_X)
Var_X = cal_Variance(X, Prop_X)
print("V(X) = ", Var_X)

# 확률변수 Y의 기댓값, 분산 구하기
Exp_Y = cal_Expected_val(Y, Prop_Y)
print("E(Y) = ", Exp_Y)
Var_Y = cal_Variance(Y, Prop_Y)
print("V(Y) = ", Var_Y)

# E(XY), Cov(X, Y), V(X+Y) 구하기
Exp_X_Y = cal_Expected_value_X_Y(X, Y, Prop_X_Y)
print("E(XY) = ", Exp_X_Y)
Cov_X_Y = cal_Corvariance_X_Y(X, Y, Prop_X_Y)
print("Cov(X, Y) = ", Cov_X_Y)
Var_X_Y = cal_joint_variance_X_Y(X, Y, Prop_X_Y)
print("V(X+Y) = ", Var_X_Y)
print("\n")

# 문제 (b)
print("문제 (b)")
X = np.array([0, 1, 2, 3])
Y = np.array([0, 1, 2, 3, 4])
Prop_X_Y = np.array([[1/20, 1/20, 3/40, 1/40],[3/40, 1/20, 1/40, 1/20],[1/40, 3/40, 1/20, 1/20], [1/20, 1/40, 3/40, 1/20],[1/20, 1/40, 1/20, 3/40]])
Prop_X = np.sum(Prop_X_Y, axis=0)
Prop_Y = np.sum(Prop_X_Y, axis=1)

# 확률변수 X의 기댓값, 분산 구하기
Exp_X = cal_Expected_val(X, Prop_X)
print("E(X) = ", Exp_X)
Var_X = cal_Variance(X, Prop_X)
print("V(X) = ", Var_X)

# 확률변수 Y의 기댓값, 분산 구하기
Exp_Y = cal_Expected_val(Y, Prop_Y)
print("E(Y) = ", Exp_Y)
Var_Y = cal_Variance(Y, Prop_Y)
print("V(Y) = ", Var_Y)

# E(XY), Cov(X, Y), V(X+Y) 구하기
Exp_X_Y = cal_Expected_value_X_Y(X, Y, Prop_X_Y)
print("E(XY) = ", Exp_X_Y)
Cov_X_Y = cal_Corvariance_X_Y(X, Y, Prop_X_Y)
print("Cov(X, Y) = ", Cov_X_Y)
Var_X_Y = cal_joint_variance_X_Y(X, Y, Prop_X_Y)
print("V(X+Y) = ", Var_X_Y)
```

```python
import numpy as np

# 데이터를 행렬 형태로 변환하는 함수
def make_matrix_form(input_X):
    matrix_form_A = []
    m = input_X.shape[0]
    n = input_X.shape[1]
    for i in range(m):
        A_row = [1]
        for j in range(n):
            num_app = input_X[i][j]
            A_row.append(num_app)
        matrix_form_A.append(A_row)
    np_matrix_form_A = np.array(matrix_form_A)
    return np_matrix_form_A

# 선형회귀식을 계산하는 함수
def cal_linear_reg_coeff(input_X, y_val):
    matrix_X = make_matrix_form(input_X)
    matrix_X_t_matrix_X = np.dot(matrix_X.T,matrix_X )
    XtX_inv = np.linalg.inv(matrix_X_t_matrix_X)
    lin_reg_coeff = np.dot(np.dot(XtX_inv, matrix_X.T),y_val)
    return lin_reg_coeff

# 문제 (a)
print("문제 (a)")
X = np.array([[2],[4],[6],[8],[10]])
b = np.array([4, 4, 3, 2, 1])

# 선형회귀식 구하기
Beta = cal_linear_reg_coeff(X, b)
print("beta = ", Beta)

# 선형회귀식 출력하기
x = sym.Symbol('x')
lin_reg = Beta[0]+Beta[1]*x
print("선형회귀식 = ", lin_reg)

# x=7일 때, 선형회귀함수의 값 출력하기
lin_reg_val = lin_reg.subs([(x, 7)])
print("x=7일 때의 값 = ", lin_reg_val)
print("\n")

# 문제 (b)
print("문제 (b)")
X = np.array([[0, 1],[1, 3],[2, 4],[3, 5],[5, 7]])
b = np.array([2, 3, 4, 6, 7])

# 선형회귀식 구하기
Beta = cal_linear_reg_coeff(X, b)
print("beta = ", Beta)

# 선형회귀식 출력하기
x, y = sym.symbols('x y')
lin_reg = Beta[0]+Beta[1]*x + Beta[2]*y
print("선형회귀식 = ", lin_reg)

# (x, y)=(3, 7)일 때, 선형회귀함수의 값 출력하기
lin_reg_val = lin_reg.subs([(x, 3), (y, 7)])
print("(3, 7)일 때의 값 = ", lin_reg_val)
```

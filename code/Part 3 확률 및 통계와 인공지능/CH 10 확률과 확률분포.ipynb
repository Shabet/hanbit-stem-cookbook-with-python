# 10.1 조건부 확률과 베이즈 정리 

예제 10-1

```python
import sympy as sym

# 동전 두 개를 던질 때 일어나는 사건
coin_list_1 = ['H','T']
coin_list_2 = ['H','T']

# 표본공간 생성
S_Space = list()
for i in range(len(coin_list_1)):
    for j in range(len(coin_list_2)):
        S_Space.append((coin_list_1[i],coin_list_2[j]))

# 두 동전 중 적어도 하나가 앞면이 나올 사건
E_Space = list()
for i in range(len(coin_list_1)):
    if coin_list_1[i] == 'H':
        for j in range(len(coin_list_2)):
            E_Space.append((coin_list_1[i],coin_list_2[j]))
    else :
        for j in range(len(coin_list_2)):
            if coin_list_2[j] == 'H':
                E_Space.append((coin_list_1[i],coin_list_2[j]))

Sample_space = sym.FiniteSet(*S_Space)
print("표본공간 S는 ? ", Sample_space)
Event = sym.FiniteSet(*E_Space)
print("사건 A는 ? ", Event)
probability = len(Event) / len(Sample_space)
print("확률은 ?", probability)
```

```python
import sympy as sym

# 동전 두 개를 던질 때 일어나는 사건
coin_list_1 = ['H','T']
coin_list_2 = ['H','T']

# 표본공간 생성
S_Space = list()
for i in range(len(coin_list_1)):
    for j in range(len(coin_list_2)):
        S_Space.append((coin_list_1[i],coin_list_2[j]))

# 두 동전 중 적어도 하나가 앞면이 나올 사건
A_Space = list()
for i in range(len(coin_list_1)):
    if coin_list_1[i] == 'H':
        for j in range(len(coin_list_2)):
            A_Space.append((coin_list_1[i],coin_list_2[j]))
    else :
        for j in range(len(coin_list_2)):
            if coin_list_2[j] == 'H':
                A_Space.append((coin_list_1[i],coin_list_2[j]))

# 두 동전이 모두 앞면이 나올 사건
B_Space = list()
for i in range(len(coin_list_1)):
    for j in range(len(coin_list_2)):
        if ((coin_list_1[i] == 'H') & (coin_list_2[j] == 'H')) == True :
            B_Space.append((coin_list_1[i],coin_list_2[j]))

Sample_space = sym.FiniteSet(*S_Space)
print("표본공간 S는 ? ", Sample_space)
Event_A = sym.FiniteSet(*A_Space)
print("사건 A는 ? ", Event_A)
Event_B = sym.FiniteSet(*B_Space)
print("사건 B는 ? ", Event_B)
A_interset_B = Event_A.intersect(Event_B)
print("사건 A와 사건 B의 교집합은 ? ", A_interset_B)
probability_B_A = (len(A_interset_B) / len(Sample_space))/(len(Event_A )/len(Sample_space))
print("사건 A가 일어났을 때, 사건 B가 일어날 확률은 ?", probability_B_A)
```

```python
import sympy as sym

# 주사위를 던질 때 일어나는 사건
dice_list_1 = range(1,7)
dice_list_2 = range(1,7)

# 표본공간 생성
S_Space = list()
for i in range(len(dice_list_1)):
    for j in range(len(dice_list_2)):
        S_Space.append((dice_list_1[i],dice_list_2[j]))

# 합이 짝수인 사건
A_Space = list()
for i in range(len(dice_list_1)):
    for j in range(len(dice_list_2)):
        if (dice_list_1[i]+dice_list_2[j])%2 == 0:
            A_Space.append((dice_list_1[i],dice_list_1[j]))
            
# 합이 3의 배수인 사건
B_Space = list()
for i in range(len(dice_list_1)):
    for j in range(len(dice_list_2)):
        if (dice_list_1[i]+dice_list_2[j])%3 == 0:
            B_Space.append((dice_list_1[i],dice_list_1[j]))
            
Sample_space = sym.FiniteSet(*S_Space)
print("표본공간 S는 ? ", Sample_space)
Event_A = sym.FiniteSet(*A_Space)
print("사건 A는 ? ", Event_A)
Event_B = sym.FiniteSet(*B_Space)
print("사건 B는 ? ", Event_B)
probability_A = len(Event_A) / len(Sample_space)
print("\n")

print("사건 A의 확률은 ?", probability_A)
probability_B = len(Event_B) / len(Sample_space)
print("사건 B의 확률은 ?", probability_B)
A_interset_B = Event_A.intersect(Event_B)
print("사건 A와 사건 B의 교집합은 ? ", A_interset_B)
probability_A_B = len(A_interset_B) / len(Sample_space)
print("사건 A와 사건 B의 교집합의 확률은 ?", probability_A_B)
print("사건 A, B는 독립인가 ? ", probability_A * probability_B == probability_A_B)
```

예제 10-5

```python
import sympy as sym

# 소수인지 확인하는 함수
def is_prime_num(num):
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

# 주사위를 던질 때 일어나는 사건
dice_list_1 = range(1,7)
dice_list_2 = range(1,7)

# 표본공간 생성
S_Space = list()
for i in range(len(dice_list_1)):
    for j in range(len(dice_list_2)):
        S_Space.append((dice_list_1[i],dice_list_2[j]))

# 합이 홀수인 사건
A_Space = list()
for i in range(len(dice_list_1)):
    for j in range(len(dice_list_2)):
        if (dice_list_1[i]+dice_list_2[j])%2 == 1:
            A_Space.append((dice_list_1[i],dice_list_1[j]))
            
# 합이 소수인 사건
B_Space = list()
for i in range(len(dice_list_1)):
    for j in range(len(dice_list_2)):
        if (dice_list_1[i]+dice_list_2[j]) == 2:
            B_Space.append((dice_list_1[i],dice_list_1[j]))
        if is_prime_num(dice_list_1[i]+dice_list_2[j]) == True:
            B_Space.append((dice_list_1[i],dice_list_1[j]))

Sample_space = sym.FiniteSet(*S_Space)
print("표본공간 S는 ? ", Sample_space)
Event_A = sym.FiniteSet(*A_Space)
print("사건 A는 ? ", Event_A)
Event_B = sym.FiniteSet(*B_Space)
print("사건 B는 ? ", Event_B)
print("\n")

probability_A = len(Event_A) / len(Sample_space)
print("사건 A의 확률은 ?", probability_A)
probability_B = len(Event_B) / len(Sample_space)
print("사건 B의 확률은 ?", probability_B)
A_interset_B = Event_A.intersect(Event_B)
print("사건 A와 사건 B의 교집합은 ? ", A_interset_B)
probability_A_B = len(A_interset_B) / len(Sample_space)
print("사건 A와 사건 B의 교집합의 확률은 ?", probability_A_B)
print("사건 A, B는 독립인가 ? ", probability_A * probability_B == probability_A_B)
probability_A_of_B = probability_A_B / probability_A
print("사건 B가 일어났을 때, 사건 A가 일어날 확률은 ?", probability_A_of_B )
print("\n")

probability_B_of_A_bayes = (probability_A_of_B * probability_A )/probability_B
print("베이즈 정리를 이용해 계산한 사건 A가 일어났을 때, 사건 B가 일어날 확률은 ?",
probability_B_of_A_bayes )
probability_B_of_A = probability_A_B / probability_B
print("사건 A가 일어났을 때, 사건 B가 일어날 확률은 ?", probability_B_of_A )
print("위의 두 확률이 같은가?" , probability_B_of_A_bayes == probability_B_of_A)
```

# 10.2 이산확률분포 

예제 10-6

```python
import scipy.stats as stats
import sympy as sym

# 동전을 던질 때 일어나는 사건
coin_list = ['H','T']

# 표본공간 생성
S_Space = list()
for i in range(len(coin_list)):
    S_Space.append(coin_list[i])
        
# 앞면이 나올 사건
E_Space = list()
for i in range(len(coin_list)):
    if coin_list[i] == 'H':
        E_Space.append(coin_list[i])
        
Sample_space = sym.FiniteSet(*S_Space)
print("표본공간 S는 ? ", Sample_space)
Event = sym.FiniteSet(*E_Space)
print("사건 A는 ? ", Event)
probability_A = len(Event) / len(Sample_space)
print("사건 A의 확률은 ?", probability_A)

# 이항분포 B(100, 1/2)
Number_of_repeats = 100
Binomial_coin = stats.binom(n = Number_of_repeats, p = probability_A)
print("\n")
print("기댓값 = ", Binomial_coin.mean())
print("분산 = ", Binomial_coin.var())
print("표준편차 = ", Binomial_coin.std())
```

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

# 이항분포 B(10, 1/2) 생성
Binomial_1 = binom(n = 10, p = 0.5)
print("기댓값 = ", Binomial_1.mean())
print("분산 = ", Binomial_1.var())
print("표준편차 = ", Binomial_1.std())
print("\n")

# 이항분포 B(20, 1/2) 생성
Binomial_2 = binom(n = 20, p = 0.5)
print("기댓값 = ", Binomial_2.mean())
print("분산 = ", Binomial_2.var())
print("표준편차 = ", Binomial_2.std())
print("\n")

# 이항분포 B(30, 1/10) 생성
Binomial_3 = binom(n = 30, p = 0.1)
print("기댓값 = ", Binomial_3.mean())
print("분산 = ", Binomial_3.var())
print("표준편차 = ", Binomial_3.std())

# 이항분포의 확률질량함수 계산
x_1 = np.arange(-3, 10.1)
y_1_pmf = Binomial_1.pmf(x_1)
x_2 = np.arange(-3, 20.1)
y_2_pmf = Binomial_2.pmf(x_2)
x_3 = np.arange(-3, 30.1)
y_3_pmf = Binomial_3.pmf(x_3)

# 이항분포의 확률질량함수의 그래프 그리기
plt.figure(figsize=(12, 8))
plt.plot(x_1, y_1_pmf , 'bo--', label="(n = 10, p = 0.5)")
plt.plot(x_2, y_2_pmf , 'go--', label="(n = 20, p = 0.5)")
plt.plot(x_3, y_3_pmf , 'ro--', label="(n = 30, p = 0.1)")
plt.xlabel('X')
plt.ylabel('P(X=x)')
plt.title('binomial distribution')
plt.legend()
plt.grid()
plt.show()

# 이항분포의 누적분포함수 계산
y_1_cdf = Binomial_1.cdf(x_1)
y_2_cdf = Binomial_2.cdf(x_2)
y_3_cdf = Binomial_3.cdf(x_3)

# 이항분포의 누적분포함수의 그래프 그리기
plt.figure(figsize=(12, 8))
plt.plot(x_1, y_1_cdf, 'bo--', label="(n = 10, p = 0.5)")
plt.plot(x_2, y_2_cdf, 'go--', label="(n = 20, p = 0.5)")
plt.plot(x_3, y_3_cdf, 'ro--', label="(n = 30, p = 0.1)")
plt.xlabel('X')
plt.ylabel(r'$\Phi_{\mu, \sigma^2}(X)$')
plt.title('cumulative distribution function')
plt.legend()
plt.grid()
plt.show()
```

예제 10-8

```python
import numpy as np
import matplotlib.pyplot as plt

# 푸아송분포 함수
def Poisson_distribution(num_k , lamb):
    poss_k = ((lamb ** num_k) * np.exp(-lamb))/np.math.factorial(num_k)
    return poss_k

x = np.arange(20)
# 푸아송분포 P(1) 생성
lamb_1 = 1
poss_1 = np.array([Poisson_distribution(n, lamb_1) for n in range(len(x))])
# 푸아송분포 P(5) 생성
lamb_5 = 5
poss_5 = np.array([Poisson_distribution(n, lamb_5) for n in range(len(x))])
# 푸아송분포 P(10) 생성
lamb_10 = 10
poss_10 = np.array([Poisson_distribution(n, lamb_10) for n in range(len(x))])

# 푸아송분포의 그래프 그리기
plt.figure(figsize=(12, 8))
plt.plot(x, poss_1, 'bo--', label=r'$ \lambda = 1$')
plt.plot(x, poss_5, 'go--', label=r'$ \lambda = 5$')
plt.plot(x, poss_10, 'ro--', label=r'$ \lambda = 10$')
plt.xlabel('k')
plt.ylabel('P(x=k)')
plt.title('Poisson distribution')
plt.legend()
plt.grid()
plt.show()
```

# 10.3 연속확률분포 

예제 10-9

```python
import scipy.stats

# 평균 = 10, 표준편차 = 2인 정규분포 객체 정의
# scipy.stats.norm(평균, 표준편차) : 각 평균, 표준편차에 해당하는 정규분포 생성
norm_dist = scipy.stats.norm(10, 2)

norm_mean = norm_dist.mean() # norm_dist.mean() : 정규분포의 기댓값 출력
norm_var = norm_dist.var() # norm_dist.var() : 정규분포의 분산 출력
norm_std = norm_dist.std() # norm_dist.std() : 정규분포의 표준편차 출력
norm_cdf = norm_dist.cdf(10) # norm_dist.cdf(x) : 정규분포의 x까지의 누적함수의 값 출력

print("기댓값 = ", norm_mean)
print("분산 = ", norm_var)
print("표준편차 = ", norm_std)
print("누적함수의 값 = ", norm_cdf)
```

예제 10-10

```python
import scipy.stats
import numpy as np

n = 100
coin_p = 1/2

# 이항분포 정의
# scipy.stats.binom(n, p) : 각 n, p에 해당하는 이항분포 생성
coin_bi_dist = scipy.stats.binom(n, coin_p)

B_dist_mean = coin_bi_dist.mean() # coin_bi_dist.mean() : 이항분포의 기댓값 출력
B_dist_std = coin_bi_dist.std() # coin_bi_dist.std() : 이항분포의 표준편차 출력

print("이항분포의 기댓값 = ", B_dist_mean)
print("이항분포의 표준편차 = ", B_dist_std)
print('\n')

# 정규분포 정의
coin_norm_dist = scipy.stats.norm(B_dist_mean, B_dist_std)

# 45~55까지의 확률
Part_1 = [45, 55]
print("45~55까지의 확률")
# 이항분포로 계산
# coin_bi_dist.cdf(x) : 이항분포의 x까지의 누적함수의 값 출력
Bi_prop_coin_45_55 = coin_bi_dist.cdf(Part_1[1]) - coin_bi_dist.cdf(Part_1[0])
print("이항분포로 구한 확률 = ", format(round(Bi_prop_coin_45_55,2), '.1%'))
# 정규분포로 계산
Norm_prop_coin_45_55 = coin_norm_dist.cdf(Part_1[1]) - coin_norm_dist.cdf(Part_1[0])
print("정규분포로 구한 확률 = ", format(round(Norm_prop_coin_45_55,2), '.1%'))
print('\n')

# 40~60까지의 확률
Part_2 = [40, 60]
print("40~60까지의 확률")
# 이항분포로 계산
Bi_prop_coin_40_60 = coin_bi_dist.cdf(Part_2[1]) - coin_bi_dist.cdf(Part_2[0])
print("이항분포로 구한 확률 = ", format(round(Bi_prop_coin_40_60,2), '.1%'))
# 정규분포로 계산
Norm_prop_coin_40_60 = coin_norm_dist.cdf(Part_2[1]) - coin_norm_dist.cdf(Part_2[0])
print("정규분포로 구한 확률 = ", format(round(Norm_prop_coin_40_60,2), '.1%'))
print('\n')

# 35~65까지의 확률
Part_3 = [35, 65]
print("35~65까지의 확률")
# 이항분포로 계산
Bi_prop_coin_35_65 = coin_bi_dist.cdf(Part_3[1]) - coin_bi_dist.cdf(Part_3[0])
print("이항분포로 구한 확률 = ", format(round(Bi_prop_coin_35_65,3), '.1%'))
# 정규분포로 계산
Norm_prop_coin_35_65 = coin_norm_dist.cdf(Part_3[1]) - coin_norm_dist.cdf(Part_3[0])
print("정규분포로 구한 확률 = ", format(round(Norm_prop_coin_35_65,4), '.1%'))
```

10.1 이산확률분포에 관한 다음 물음에 답하라.

```python
# 문제 (a)
import numpy as np
from scipy.special import comb
import matplotlib.pyplot as plt

# 초기하분포 H(N, n, k) 함수
def HyperGeometric_distribution(num_N, num_n, num_k, num_x):
    poss_k = (comb(num_k, num_x) * comb(num_N-num_k, num_n-num_x))/comb(num_N, num_n)
    return poss_k

x = np.arange(18)
N = [32, 32]
n = [16, 24]
k = [16, 16]

# 초기하분포 H(32,16,16) 생성
HyperGeometric_poss_1 = np.array([HyperGeometric_distribution(N[0], n[0], k[0], x_1) for
x_1 in range(len(x))])

# 초기하분포 H(32,24,16) 생성
HyperGeometric_poss_2 = np.array([HyperGeometric_distribution(N[1], n[1], k[1], x_1) for
x_1 in range(len(x))])

# 초기하분포의 그래프 그리기
plt.figure(figsize=(12, 8))
plt.plot(x, HyperGeometric_poss_1, 'bo--', label=r'$H(32,16,16)$')
plt.plot(x, HyperGeometric_poss_2, 'go--', label=r'$H(32,24,16)$')
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('hypergeometric distribution')
plt.legend()
plt.grid()
plt.show()

# 문제 (b)
import numpy as np
import matplotlib.pyplot as plt

# 기하분포 GEO(p) 함수
def Geometric_distribution(prop_p, num_x):
    poss_k = prop_p * (1-prop_p)**(num_x-1)
    return poss_k

x = np.arange(1,18)
Prop_p = [0.2, 0.5, 0.7]

# 기하분포 GEO(0.2) 생성
Geometric_poss_1 = np.array([Geometric_distribution(Prop_p[0], x_1) for x_1 in range(1,len(x)+1)])
# 기하분포 GEO(0.5) 생성
Geometric_poss_2 = np.array([Geometric_distribution(Prop_p[1], x_1) for x_1 in range(1,len(x)+1)])
# 기하분포 GEO(0.7) 생성
Geometric_poss_3 = np.array([Geometric_distribution(Prop_p[2], x_1) for x_1 in range(1,len(x)+1)])

# 기하분포의 그래프 그리기
plt.figure(figsize=(12, 8))
plt.plot(x, Geometric_poss_1, 'bo--', label=r'$GEO(0.2)$')
plt.plot(x, Geometric_poss_2, 'go--', label=r'$GEO(0.5)$')
plt.plot(x, Geometric_poss_3, 'ro--', label=r'$GEO(0.7)$')
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('geometric distribution')
plt.legend()
plt.grid()
plt.show()
```

```python
# 문제 (a)
import numpy as np
from scipy.special import gamma
import matplotlib.pyplot as plt

# 감마분포 Γ(a, b) 함수
def Gamma_distribution(a_val, b_val, num_x):
    poss_x = num_x**(a_val-1) * np.exp(-num_x/b_val) / (b_val**a_val*gamma(a_val))
    return poss_x

x = np.arange(0,50,0.4)

# 감마분포 Γ(5, 2) 생성
Gamma_poss_1 = np.array([Gamma_distribution(5, 2, x_1) for x_1 in x])
# 감마분포 Γ(10, 2) 생성
Gamma_poss_2 = np.array([Gamma_distribution(10, 2, x_2) for x_2 in x])
# 감마분포 Γ(15, 2) 생성
Gamma_poss_3 = np.array([Gamma_distribution(15, 2, x_3) for x_3 in x])
# 감마분포 Γ(20, 2) 생성
Gamma_poss_4 = np.array([Gamma_distribution(20, 2, x_4) for x_4 in x])

# 감마분포의 그래프 그리기
plt.figure(figsize=(12, 8))
plt.plot(x, Gamma_poss_1, 'bo--', label=r'$Gamma(5,2)$')
plt.plot(x, Gamma_poss_2, 'yo--', label=r'$Gamma(10,2)$')
plt.plot(x, Gamma_poss_3, 'go--', label=r'$Gamma(15,2)$')
plt.plot(x, Gamma_poss_4, 'ro--', label=r'$Gamma(20,2)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('gamma distribution')
plt.legend()
plt.grid()
plt.show()

# 문제 (b)
import numpy as np
from scipy.special import gamma
import matplotlib.pyplot as plt

# 지수분포 Exp(theta) 함수
def Exponential_distribution(theta_val, num_x):
    poss_x = (1/theta_val) * np.exp(-num_x/theta_val)
    return poss_x

x = np.arange(0,5,0.1)

# 지수분포 Exp(0.5) 생성
Exponential_poss_1 = np.array([Exponential_distribution(0.5, x_1) for x_1 in x])
# 지수분포 Exp(1) 생성
Exponential_poss_2 = np.array([Exponential_distribution(1, x_2) for x_2 in x])
# 지수분포 Exp(2) 생성
Exponential_poss_3 = np.array([Exponential_distribution(2, x_3) for x_3 in x])

# 지수분포의 그래프 그리기
plt.figure(figsize=(12, 8))
plt.plot(x, Exponential_poss_1, 'bo--', label=r'$Exp(0.5)$')
plt.plot(x, Exponential_poss_2, 'yo--', label=r'$Exp(1)$')
plt.plot(x, Exponential_poss_3, 'go--', label=r'$Exp(2)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('exponential distribution')
plt.legend()
plt.grid()
plt.show()
```
